```python
def plot(img):
    plt.figure(1)
    plt.imshow(img, cmap='Greys_r')
    
def comparePlot(i1, i2):
    fig = plt.figure()
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    ax1.imshow(i1, cmap='Greys_r')
    ax2.imshow(i2, cmap='Greys_r')

plot(original_image[:,:,0])
```

```python

def w_frac(i, j, gi, gj, alpha_s, alpha_r):
    imj = i - j
    gimgj = gi - gj
    return 1 / ((np.dot(imj, imj) ** (alpha_s / 2) + 0.0001) * (np.dot(gimgj, gimgj) ** (alpha_r / 2) + 0.0001))

def w_exp(i, j, gi, gj, sigma_s, sigma_r):
    imj = i - j
    gimgj = gi - gj
    return np.e**(- np.dot(imj, imj) / (2 * sigma_s**2)) * np.e(- np.dot(gimgj, gimgj) / (2 * sigma_r**2))
```

```python
def buildA(img, lambda_, r, par1, par2, useFracWeights=True):

    m , n = img.shape[0], img.shape[1] 
    size = m * n

    rowInd = []
    colInd = []
    data = []


    for i in range(m):
        mink = max(i - r, 0)
        for j in range(n):
            minl = max(j - r, 0)
            I = i * n + j
            tot = 0
            for k in range(mink, min(i + r + 1, m)):
                for l in range(minl, min(j + r + 1, n)):
                    J = k * n + l
                    if I != J:
                        if (useFracWeights):
                            val = lambda_ * w_frac(np.array([i, j]), np.array([k, l]), 
                                           original_image[i, j], original_image[k, l], par1, par2)
                        else:
                            val = lambda_ * w_exp(np.array([i, j]), np.array([k, l]), 
                                           original_image[i, j], original_image[k, l], par1, par2)
                        tot += val
                        rowInd.append(I)
                        colInd.append(J)
                        data.append(-val)
            rowInd.append(I)
            colInd.append(I)
            data.append(tot + 1)
        if i % 100 == 0 : print(i)
   
    return sp.coo_matrix((data, (rowInd, colInd)), shape=(size, size))
```

```python
lambdas = [5, 40, 320]
Us = []
r = 1
alpha_r = alpha_s = 1.2
for lambda_ in lambdas:
    A = buildA(original_image[:,:,0], lambda_, r, alpha_s, alpha_r)
    U = sp.linalg.spsolve(A, original_image[:,:,0].flatten())
    Us.append(U)
    plot(U.reshape((m, n)))
```
